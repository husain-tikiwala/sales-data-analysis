```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
titan=pd.read_csv("Titanic-Dataset.csv")
```

```python
titan
```

```python
titan=titan.drop("Cabin",axis=1)
```

```python
titan
```

```python
sns.countplot(x="Sex",data=titan)
```

```python
titan['Age'].hist(bins= 20)
plt.title('Age distribution')
plt.xlabel('Age')
plt.ylabel('Frquency')
plt.show()
```

```python
sns.kdeplot(x="Fare",data=titan)
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Density')
plt.show()
```

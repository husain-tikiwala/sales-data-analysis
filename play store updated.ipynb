```python
plt.bar(play["Category"].unique(),play.groupby("Category")["Installs"].sum(),color=("green"))
plt.xlabel("category")
plt.ylabel("no. of installs")
plt.title("category vs installs",color="blue")
plt.xticks(size=5,rotation=90)
plt.yticks(size=14,rotation=60)
plt.show()
```
